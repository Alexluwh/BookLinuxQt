# 需求分析

分层架构图

## 医生登录

医生工作站程序在登录的时候填写手机号和密码，并且发送到服务器。
服务器查询数据库，如果手机号查询不到，说明用户未注册。
如果手机号和密码都匹配，则允许用户登录。

数据库表保存每个医生的记录，包含手机号和密码两个字段。

## 病人列表

在表格显示所有病人，支持增删改查。

数据库添加病人表，字段包括：姓名、性别、年龄。
因为每年的年龄都需要加一，所以不容易维护，一般不设年龄为字段，改为设置生日字段，
这样只需要用当前的日期减去生日，就能够轻松求得年龄。


## 设备列表

设备需要带有一个序列号，以便识别设备的身份。
这次实现简化需求，设备第一次登录系统的时候在数据库增加一条记录。

设备如果不在线，应该及时提醒医护人员处理。

如何知道在线状态？采用保活报文的机制。
因为互联网是无连接的，无法知道电路连接的状态，
设备端每隔一定周期（15秒）发送保活报文（定时器），
记录每次保活的时间，如果超过20秒，认为离线。


## 设备与病人绑定

医生在工作站新建病人记录。
在病人列表，可以选择关联机器；在机器列表，也可以选择关联病人。

一台设备只可以给一名病人使用，但是使用完毕会切换给其他病人使用。
为了记录使用日志，绑定的关系需完整如实保存，所以需要建一个专门的表。

## 历史数据
心率每分钟，统计最大、最小和平均值，发送到数据库保存。
